# Challenge: Space Challenges

Reference Link: https://github.com/rfordatascience/tidytuesday/tree/master/data/2019/2019-01-15

```python
import pandas as pd 
import numpy as np

import os, ssl
if (not os.environ.get('PYTHONHTTPSVERIFY', '') and getattr(ssl, '_create_unverified_context', None)):
    ssl._create_default_https_context = ssl._create_unverified_context

launches_df = (
    pd.read_csv('https://raw.githubusercontent.com/rfordatascience/' + \
                'tidytuesday/master/data/2019/2019-01-15/launches.csv')
    .convert_dtypes()
)

launches_df.info()
```

1) Initial Data Setup
  + The Soviet Union was renamed to Russia partway through the dataset.In the state_code column change all occurrences of "SU" to "RU"
  +  Exclude agency_type "private", "startup" from the dataset
  +  Add a column called launch_month that extracts the month from the launch_date column

```python
clean_df = (
    launches_df
    .assign(state_code = lambda x: np.where(x.state_code == "SU", "RU", x.state_code),
            launch_date = lambda x: x.launch_date.str.replace('2918', '2018').pipe(pd.to_datetime),
            launch_month = lambda x: x.launch_date.dt.month.astype('Int64'))
    [lambda x: ~x.agency_type.isin(['private', 'startup'])]
    .convert_dtypes()
    [['state_code', 'launch_year', 'launch_month', 'category']]
)

clean_df.head()
```

2) Who are the top 4 countries (state_code) in space launch count?
  +  Create a data frame that shows each country, their respective count of launches, and their rank

```python
top_df = (
    clean_df
    .groupby('state_code', as_index = False)
    .agg(n = ('launch_year', 'count'))
    .sort_values('n', ascending = False)
    .head(4)
    .assign(rank = lambda x: np.arange(len(x)) + 1)
)

top_df
```

3) For the top two countries (state_code) what is their success rate by decade
  +  Note: The category column marks whether a launch is successful. A launch is a success if category is "O" and unsuccessful if category is "F".
  + Success rate by decade is defined as the percentage of successful launches over all the launches performed by a particular country in a decade.

```python
top_list = top_df.head(2).state_code

success_df = (
    clean_df
    [lambda x: x.state_code.isin(top_list)]
    .assign(launch_decade = lambda x: x.launch_year.floordiv(10).multiply(10))
    .groupby(['state_code', 'launch_decade', 'category'], as_index = False)
    .agg(n = ('launch_year', 'count'))
    .assign(success_pct = lambda x: x.n/x.groupby(['state_code', 'launch_decade'])['n'].transform('sum'))
    [lambda x: x.category == "O"]
    .drop(['category', 'n'], 1)
    .round(4)
)

success_df
```
